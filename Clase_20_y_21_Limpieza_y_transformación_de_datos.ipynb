## ¿Cómo podemos tratar los valores atípicos?

````python
import seaborn as sns
import matplotlib.pyplot as plt
````

````python
# Crear la figura y los subplots
fig, axs = plt.subplots(1, 3, figsize=(10, 4))

# Graficar el primer histograma en el primer subplot
sns.histplot(data=data, x='income', ax=axs[0])

# Configurar título y etiquetas del primer subplot
axs[0].set_title('Histograma 1')
axs[0].set_xlabel('Income')
axs[0].set_ylabel('Frecuencia')

# Graficar la distribución en el segundo subplot
sns.kdeplot(data=data, x='income', ax=axs[1])

# Configurar título y etiquetas del segundo subplot
axs[1].set_title('Distribución')
axs[1].set_xlabel('income')
axs[1].set_ylabel('Frecuencia')

# Graficar el boxplot en el tercer subplot
sns.boxplot(data=data, x='income', ax=axs[2])

# Configurar título y etiquetas del segundo subplot
axs[2].set_title('Boxplot')
axs[2].set_xlabel('income')
axs[2].set_ylabel('Frecuencia')

# Ajustar el espaciado entre subplots
plt.tight_layout()

# Mostrar la figura con los subplots
plt.show()
````
