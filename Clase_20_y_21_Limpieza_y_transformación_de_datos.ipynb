## Filtrando los datos para eliminar el ruido


```python
import pandas as pd
```

```python
data = pd.read_csv('employee.csv')
data
```

Tenemos varias formas de seleccionar una parte de nuestro dataframe o filtrar nuestro dataframe

**1. filter**

```python
data.filter(['name','department'])
```

**2. lista de listas (para varias columnas) o lista (para una sola columna)**

```python
data[['name','department']]
```

**3. Con los índices usando loc, and iloc**



```python
data.loc[:,['name','department']]
```

```python
data.loc[:,'age':'grade']
```

```python
#índice implicito
data.iloc[:,1:-2]
```
