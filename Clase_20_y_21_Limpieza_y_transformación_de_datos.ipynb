## ¿Cómo podemos tratar los valores faltantes?


**1. Podemos eliminarlos**

````python
#Creamos una copia del dataframe para no afectar el original
data_1 = data.copy()
data_1
````

````python
# Eliminamos los nulos
data_1.dropna()
````

````python
data_1
````

````python
#Eliminamos los nulos reemplazando la variable
#data_1 = data_1.dropna()
data_1.dropna(inplace=True)
````

````python
data_1
````

**2. Llenar de valores faltantes de la columna mediana**


````python
#Creamos una copia del dataframe para no afectar el original
data_2 = data.copy()
````

````python
data_2
````

````python
#Creamos la mediana
mediana = data_2.age.median()
mediana
````

````python
#LLenamos los nulos con la mediana con fillna
#data_2['age'] = data_2['age'].fillna(data_2.age.median())
data_2['age'] = data_2['age'].fillna(mediana)
````

````python
promedio = data_2.income.mean()
promedio
````

````python
data_2['income'] = data_2['income'].fillna(promedio)
````

## ¿Cómo podemos tratar los valores atípicos?

````python
import seaborn as sns
import matplotlib.pyplot as plt
````

````python
# Crear la figura y los subplots
fig, axs = plt.subplots(1, 3, figsize=(10, 4))

# Graficar el primer histograma en el primer subplot
sns.histplot(data=data, x='income', ax=axs[0])

# Configurar título y etiquetas del primer subplot
axs[0].set_title('Histograma 1')

# Graficar la distribución en el segundo subplot
sns.kdeplot(data=data, x='income', ax=axs[1])

# Configurar título y etiquetas del segundo subplot
axs[1].set_title('Distribución')

# Graficar el boxplot en el tercer subplot
sns.boxplot(data=data, x='income', ax=axs[2])

# Configurar título y etiquetas del segundo subplot
axs[2].set_title('Boxplot')

# Ajustar el espaciado entre subplots
plt.tight_layout()

# Mostrar la figura con los subplots
plt.show()
````
