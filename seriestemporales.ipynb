# COMBINACIÓN Y FUSIÓN DE CONJUNTOS DE DATOS

```python
data1 = pd.read_csv('data1.csv')
data1
```

```python
data2 = pd.read_csv('data2.csv')
data2
```

```python
datac= pd.concat([data1,data2], ignore_index=True)
datac
```

```python
venta =pd.read_csv('venta.csv')
venta
```

```python
productomerge = pd.merge(datac,venta, on=['ID','Nombre'])
productomerge
```

```python
cliente = pd.read_csv('cliente.csv')
cliente
```

```python
total = pd.merge(productomerge,cliente, on =['Cliente_ID'], suffixes= ('_Producto','_Cliente'))
total
```

```python
inicial = '2022-01-01'
final='2022-01-10'

fechas_diario = pd.date_range(start=inicial,end =final,freq ='D')

datos_diario = [50,45,60,55,75,80,65,70,90,85]

df_diario = pd.DataFrame({'Fecha':fechas_diario,'Datos': datos_diario})

df_diario.set_index('Fecha', inplace=True)

df_diario
```

```python
#Resamplea los datos para obtener promedios semanales

semanas = df_diario.resample('W').mean()
semanas
```

```python
#Calcula la diferencia en días entre la primera y la última fecha

intervalo = df_diario.index.max() - df_diario.index.min()
intervalo
```
