```python
fechas_ventas_temporales = pd.date_range(start='2022-01-01', end='2022-12-31', freq='D') 
ventas_temporales = np.random.randint(50, 200, size=len(fechas_ventas_temporales)) 
df_ventas_temporales = pd.DataFrame({'Fecha': fechas_ventas_temporales, 'Ventas': ventas_temporales})
```

```python
df_ventas_temporales['Mes'] = df_ventas_temporales['Fecha'].dt.month
```

```python
df_ventas_temporales
```

```python
df_ventas_temporales.set_index('Fecha', inplace = True)
df_ventas_temporales
```

```python
df_ventas_mes = df_ventas_temporales.resample('M').mean()
df_ventas_mes
```

```python
promedio_ventas_por_mes = df_ventas_temporales.groupby('Mes')['Ventas'].mean()
```

```python
promedio_ventas_por_mes
```

''' Combinación de Dataframes
Enunciado
Dado el Dataframe df_productos:
datos_productos = {'Producto': ['A', 'B', 'C'], 'Precio_Unitario': [10, 20, 30]}
df_productos = pd.DataFrame(datos_productos)
y el Dataframe df_ventas_productos:
datos_ventas_productos = {'Producto': ['A', 'B', 'C'], 'Cantidad': [50, 30, 40]}
df_ventas_productos = pd.DataFrame(datos_ventas_productos)
a) Combina ambos DataFrames para obtener información completa sobre las ventas de productos, incluyendo el precio unitario.
b) Calcula el monto total de ventas para cada producto '''

```python
datos_productos = {'Producto': ['A', 'B', 'C'], 'Precio_Unitario': [10, 20, 30]}
```

```python
df_productos = pd.DataFrame(datos_productos)
```

```python
datos_ventas_productos = {'Producto': ['A', 'B', 'C'], 'Cantidad': [50, 30, 40]}
```

```python
df_ventas_productos = pd.DataFrame(datos_ventas_productos)
```

```python
df_completo = pd.merge(df_productos,df_ventas_productos, on = ['Producto'])
df_completo
```

```python
df_completo['Total'] = df_completo['Precio_Unitario'] * df_completo['Cantidad']
df_completo
```
