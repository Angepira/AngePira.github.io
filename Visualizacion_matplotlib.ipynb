El parámetro startangle de la función pie() en Matplotlib se utiliza para especificar el ángulo inicial de la gráfica de torta. Por defecto, el ángulo inicial es cero, lo que significa que el primer segmento de la torta se dibuja a las 3 en punto en el sentido de las agujas del reloj.

Al cambiar el valor del parámetro startangle, se puede rotar la gráfica de torta en sentido horario o antihorario para comenzar en un ángulo diferente. Por ejemplo, si se establece startangle=90, el primer segmento se dibuja a las 12 en punto, y la gráfica de torta se gira en sentido antihorario.
