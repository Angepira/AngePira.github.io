```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
data={'Ciudad':['Bogotá','Medellin','Cartagena','Cali','Santa Marta'],
      'Mujeres':[15000,12000,9000,8000,7000],
      'Hombres':[18000,11000,10000,7500,8000]}
df_turismo=pd.DataFrame(data)
df_turismo['Total']=df_turismo['Mujeres']+df_turismo['Hombres']
df_turismo
```

```python
#Grafico tipo torta
plt.pie(df_turismo['Total'],labels=df_turismo['Ciudad'], autopct="%1.1f%%" , startangle=140)
plt.title('Visitas Turisticas por Ciudad en 2023')
plt.show()
```

```python
x=df_turismo['Ciudad']
mujeres=df_turismo['Mujeres']
hombres=df_turismo['Hombres']
plt.bar(x,hombres, color='b', label='Hombres')
plt.bar(x,mujeres, bottom=hombres, color='pink', label='Mujeres')
plt.legend()
plt.title("Visitas Turisticas por Genero en el 2023")
plt.xlabel("Ciudad")
plt.ylabel("Visitantes")
plt.minorticks_on()
plt.show()
```

```python
plt.hist(df_turismo['Hombres'],edgecolor='black', bins=10)
plt.show()
```
